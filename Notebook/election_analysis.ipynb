### REGIONAL ANALYSIS

#### Aggregate seat counts region-wise for alliance-level comparison

```python
# =========================================================
# STEP 1 — CREATE ALLIANCE COLUMN FOR 2021
# =========================================================

alliance_2021_map = {

    # DMK Alliance
    'DMK': 'DMK+',
    'CPI': 'DMK+',
    'CPI(M)': 'DMK+',
    'INC': 'DMK+',
    'VCK': 'DMK+',

    # AIADMK Alliance
    'AIADMK': 'AIADMK+',
    'BJP': 'AIADMK+',
    'PMK': 'AIADMK+',

    # NTK
    'NTK': 'NTK'
}

# Create alliance column
winner_2021['alliance'] = (
    winner_2021['winning_party_2021']
    .map(alliance_2021_map)
    .fillna('Others')
)

# =========================================================
# STEP 2 — CREATE ALLIANCE COLUMN FOR 2026
# =========================================================

alliance_2026_map = {

    # DMK Alliance
    'DMK': 'DMK+',
    'CPI': 'DMK+',
    'CPI(M)': 'DMK+',
    'DMDK': 'DMK+',
    'INC': 'DMK+',
    'IUML': 'DMK+',
    'VCK': 'DMK+',

    # AIADMK Alliance
    'AIADMK': 'AIADMK+',
    'AMMK': 'AIADMK+',
    'BJP': 'AIADMK+',
    'PMK': 'AIADMK+',

    # TVK
    'TVK': 'TVK',

    # NTK
    'NTK': 'NTK'
}

# Create alliance column
winner_2026['alliance'] = (
    winner_2026['winning_party_2026']
    .map(alliance_2026_map)
    .fillna('Others')
)

# =========================================================
# STEP 3 — REGION-WISE SEAT COUNT (2021)
# =========================================================

region_seats_2021 = (
    winner_2021
    .groupby(['region', 'alliance'])
    .size()
    .reset_index(name='no_seats_2021')
)

# =========================================================
# STEP 4 — REGION-WISE SEAT COUNT (2026)
# =========================================================

region_seats_2026 = (
    winner_2026
    .groupby(['region', 'alliance'])
    .size()
    .reset_index(name='no_seats_2026')
)

# =========================================================
# STEP 5 — MERGE BOTH YEARS
# =========================================================

region_seat_comparison = pd.merge(
    region_seats_2021,
    region_seats_2026,
    on=['region', 'alliance'],
    how='outer'
)

# =========================================================
# STEP 6 — FILL MISSING VALUES WITH 0
# =========================================================

region_seat_comparison['no_seats_2021'] = (
    region_seat_comparison['no_seats_2021']
    .fillna(0)
    .astype(int)
)

region_seat_comparison['no_seats_2026'] = (
    region_seat_comparison['no_seats_2026']
    .fillna(0)
    .astype(int)
)

# =========================================================
# STEP 7 — CREATE NET CHANGE COLUMN
# =========================================================

region_seat_comparison['net_change'] = (
    region_seat_comparison['no_seats_2026']
    -
    region_seat_comparison['no_seats_2021']
)

# =========================================================
# STEP 8 — SORT TABLE
# =========================================================

region_seat_comparison = region_seat_comparison.sort_values(
    ['region', 'alliance']
)
region_seat_comparison
```

### Victory Margin Calculation

#### Calculate constituency-level victory margins using winner and runner-up vote difference.

```python
# =========================================================
# FUNCTION TO CALCULATE MARGIN
# =========================================================

def calculate_margin(df, year):

    # -----------------------------------------------------
    # SORT CANDIDATES BY VOTES (HIGH TO LOW)
    # -----------------------------------------------------

    sorted_df = df.sort_values(
        ['ac_number', 'votes'],
        ascending=[True, False]
    )

    # -----------------------------------------------------
    # GET TOP 2 CANDIDATES IN EACH CONSTITUENCY
    # -----------------------------------------------------

    top2 = (
        sorted_df
        .groupby('ac_number')
        .head(2)
    ).copy()

    # -----------------------------------------------------
    # RANK CANDIDATES
    # -----------------------------------------------------

    top2['rank'] = (
        top2
        .groupby('ac_number')['votes']
        .rank(method='first', ascending=False)
    )

    # -----------------------------------------------------
    # SPLIT WINNER & RUNNER-UP
    # -----------------------------------------------------

    winner = top2[top2['rank'] == 1].copy()
    runner = top2[top2['rank'] == 2].copy()

    # -----------------------------------------------------
    # RENAME COLUMNS
    # -----------------------------------------------------

    winner = winner.rename(columns={
        'votes': f'winner_votes_{year}'
    })

    runner = runner.rename(columns={
        'votes': f'runner_votes_{year}'
    })

    # -----------------------------------------------------
    # MERGE WINNER + RUNNER-UP
    # -----------------------------------------------------

    margin_df = pd.merge(
        winner[
            ['ac_number', 'region', f'winner_votes_{year}']
        ],
        runner[
            ['ac_number', f'runner_votes_{year}']
        ],
        on='ac_number'
    )

    # -----------------------------------------------------
    # CALCULATE MARGIN
    # -----------------------------------------------------

    margin_df[f'margin_{year}'] = (
        margin_df[f'winner_votes_{year}']
        -
        margin_df[f'runner_votes_{year}']
    )

    return margin_df


# =========================================================
# CALCULATE FOR BOTH YEARS
# =========================================================

margin_2021 = calculate_margin(merged_2021, 2021)

margin_2026 = calculate_margin(merged_2026, 2026)

# =========================================================
# REGION-WISE AVERAGE MARGIN
# =========================================================

region_margin_2021 = (
    margin_2021
    .groupby('region')['margin_2021']
    .mean()
    .reset_index()
)

region_margin_2021 = region_margin_2021.rename(columns={
    'margin_2021': 'average_margin_2021'
})

region_margin_2026 = (
    margin_2026
    .groupby('region')['margin_2026']
    .mean()
    .reset_index()
)

region_margin_2026 = region_margin_2026.rename(columns={
    'margin_2026': 'average_margin_2026'
})

# =========================================================
# MERGE REGION TABLES
# =========================================================

region_margin_comparison = pd.merge(
    region_margin_2021,
    region_margin_2026,
    on='region'
)

# =========================================================
# CREATE NET CHANGE
# =========================================================

region_margin_comparison['net_change'] = (
    region_margin_comparison['average_margin_2026']
    -
    region_margin_comparison['average_margin_2021']
)

region_margin_comparison['average_margin_2021'] = (
    region_margin_comparison['average_margin_2021']
    .round()
    .astype(int)
)

region_margin_comparison['average_margin_2026'] = (
    region_margin_comparison['average_margin_2026']
    .round()
    .astype(int)
)

region_margin_comparison['net_change'] = (
    region_margin_comparison['net_change']
    .round()
    .astype(int)
)


# =========================================================
# OVERALL STATE AVERAGE MARGIN
# =========================================================

overall_avg_margin_2021 = (
    margin_2021['margin_2021']
    .mean()
)

overall_avg_margin_2026 = (
    margin_2026['margin_2026']
    .mean()
)


overall_avg_margin_2021 = int(round(overall_avg_margin_2021))
overall_avg_margin_2026 = int(round(overall_avg_margin_2026))


overall_margin_table = pd.DataFrame({
    'Year': [2021, 2026],
    'overall_average_margin': [
        overall_avg_margin_2021,
        overall_avg_margin_2026
    ]
})

# =========================================================
# CHECK OUTPUTS
# =========================================================

print("\nREGION-WISE AVERAGE MARGIN")
print(region_margin_comparison)

print("\nOVERALL AVERAGE MARGIN")
print(overall_margin_table)
```

### Winner Vote Share Calculation

 This function calculates:
 1. Total valid votes in each constituency
 2. Winning candidate in each constituency
 3. Winner vote share percentage
 4. Final constituency-level vote share table

```python
# =========================================================
# FUNCTION TO CALCULATE WINNER VOTE SHARE
# =========================================================

def winner_vote_share(df, year):

    # -----------------------------------------------------
    # TOTAL VALID VOTES IN EACH CONSTITUENCY
    # -----------------------------------------------------

    total_votes = (
        df
        .groupby('ac_number')['votes']
        .sum()
        .reset_index(name='total_valid_votes')
    )

    # -----------------------------------------------------
    # WINNING CANDIDATE
    # -----------------------------------------------------

    winners = df.loc[
        df.groupby('ac_number')['votes'].idxmax()
    ].copy()

    # -----------------------------------------------------
    # RENAME WINNER VOTES
    # -----------------------------------------------------

    winners = winners.rename(columns={
        'votes': 'winner_votes'
    })

    # -----------------------------------------------------
    # MERGE
    # -----------------------------------------------------

    vote_share = pd.merge(
        winners[
            [
                'ac_number',
                'constituency',
                'region',
                'party',
                'candidate',
                'winner_votes'
            ]
        ],
        total_votes,
        on='ac_number'
    )

    # -----------------------------------------------------
    # CALCULATE VOTE SHARE %
    # -----------------------------------------------------

    vote_share[f'vote_share_{year}'] = (
        vote_share['winner_votes']
        /
        vote_share['total_valid_votes']
    ) * 100

    # -----------------------------------------------------
    # ROUND
    # -----------------------------------------------------

    vote_share[f'vote_share_{year}'] = (
        vote_share[f'vote_share_{year}']
        .round(2)
    )

    return vote_share


# =========================================================
# CREATE DATASETS
# =========================================================

vote_share_2021 = winner_vote_share(merged_2021, 2021)

vote_share_2026 = winner_vote_share(merged_2026, 2026)

# =========================================================
# 2021 COUNTS
# =========================================================

gt_50_2021 = (
    vote_share_2021[
        vote_share_2021['vote_share_2021'] > 50
    ]
    .shape[0]
)

between_35_50_2021 = (
    vote_share_2021[
        (vote_share_2021['vote_share_2021'] >= 35)
        &
        (vote_share_2021['vote_share_2021'] <= 50)
    ]
    .shape[0]
)

lt_35_2021 = (
    vote_share_2021[
        vote_share_2021['vote_share_2021'] < 35
    ]
    .shape[0]
)

# =========================================================
# 2026 COUNTS
# =========================================================

gt_50_2026 = (
    vote_share_2026[
        vote_share_2026['vote_share_2026'] > 50
    ]
    .shape[0]
)

between_35_50_2026 = (
    vote_share_2026[
        (vote_share_2026['vote_share_2026'] >= 35)
        &
        (vote_share_2026['vote_share_2026'] <= 50)
    ]
    .shape[0]
)

lt_35_2026 = (
    vote_share_2026[
        vote_share_2026['vote_share_2026'] < 35
    ]
    .shape[0]
)

# =========================================================
# CREATE SUMMARY TABLE
# =========================================================

vote_share_summary = pd.DataFrame({

    'Year': [2021, 2026],

    '>50%_vote_share': [
        gt_50_2021,
        gt_50_2026
    ],

    '35%-50%_vote_share': [
        between_35_50_2021,
        between_35_50_2026
    ],

    '<35%_vote_share': [
        lt_35_2021,
        lt_35_2026
    ]
})

# =========================================================
# CHECK OUTPUT
# =========================================================

print(vote_share_summary)


```
